#### About Variables
- Math: a value that varies, algebra
- Most programming languages, books, courses, and culture:
  - Compiler reserves memory and issues type-specific instructions
  - Assignment changes the value in memory
  - Aliases are rare except as function arguments passed by reference
- Python:
  - Shares math definition with programming languages
  - Unlike traditional programming language definition
  - To avoid the baggage, say "names" and "objects"

Example of type-specific instructions (from https://godbolt.org):

    int square(int a) {          int square(float a) {
        return a * a;                return a * a;
    }                            }
    ----------------------       -----------------------
    pushq   %rbp                 pushq   %rbp
    movq    %rsp, %rbp           movq    %rsp, %rbp
    movl    %edi, -4(%rbp)  !=   movss   %xmm0, -4(%rbp)
    movl    -4(%rbp), %eax  !=   movss   -4(%rbp), %xmm0
    imull   -4(%rbp), %eax  !=   mulss   -4(%rbp), %xmm0
                            !=   cvttss2si %xmm0, %eax
    popq    %rbp                 popq    %rbp
    ret                          ret

    imull is signed multiply (integer)
    mulss is multiply scalar single-precision (float)
